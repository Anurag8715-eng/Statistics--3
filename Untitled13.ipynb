```python
''''

Question 1: What is hypothesis testing in statistics?
Answer: Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves formulating two competing hypotheses — the null hypothesis and the alternative hypothesis — and using sample evidence to determine which hypothesis is more likely to be true. It helps evaluate whether observed data differs significantly from what is expected.


Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
Answer: The null hypothesis (H₀) is a statement that assumes there is no effect or no difference; it represents the default or status quo assumption.
The alternative hypothesis (H₁ or Ha) is a statement that contradicts the null hypothesis and suggests that there is an effect or a difference.

Example:
If we are testing whether a new drug is more effective than the old one:
H₀: The new drug is not more effective than the old drug.
H₁: The new drug is more effective than the old drug.


Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
Answer: The significance level (α) is the probability of rejecting the null hypothesis when it is actually true — it represents the risk of making a Type I error.
Common values for α are 0.05, 0.01, or 0.10.
If the p-value (probability value) obtained from the test is less than α, we reject the null hypothesis; otherwise, we fail to reject it.
Thus, the significance level acts as a threshold for determining whether the results are statistically significant.


Question 4: What are Type I and Type II errors? Give examples of each.
Answer: A Type I error occurs when we reject a true null hypothesis (false positive).

Example: Concluding a new medicine works when it actually doesn’t.

A Type II error occurs when we fail to reject a false null hypothesis (false negative).
Example: Concluding a new medicine doesn’t work when it actually does.


Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
Answer: A Z-test is used when the population variance is known or when the sample size is large (n ≥ 30). It assumes the data follows a normal distribution.
A T-test is used when the population variance is unknown and the sample size is small (n < 30). It uses the sample standard deviation as an estimate of the population standard deviation.

''''
```

```python
# Question 6: Python program to generate a binomial
# distribution and plot histogram

import numpy as np
import matplotlib.pyplot as plt
n, p = 10, 0.5
data = np.random.binomial(n, p, 1000)
plt.hist(data, bins=range(0, n+2), align='left', rwidth=0.8)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
# Question 7: Hypothesis testing using Z-statistics

import numpy as np
from scipy import stats
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
mu_0 = 50
mean_sample = np.mean(sample_data)
std_sample = np.std(sample_data, ddof=1)
n = len(sample_data)
z_stat = (mean_sample - mu_0) / (std_sample / np.sqrt(n))
p_val = 2 * (1 - stats.norm.cdf(abs(z_stat)))
print("Z-statistic:", z_stat)
print("p-value:", p_val)
```

```python
# Question 8: Simulate normal data and compute 95%
# confidence interval

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
np.random.seed(42)
data = np.random.normal(100, 15, 500)
mean = np.mean(data)
std = np.std(data, ddof=1)
ci_low, ci_high = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std/np.sqrt(len(data)))
print("Mean:", mean)
print("95% Confidence Interval:", (ci_low, ci_high))
plt.hist(data, bins=30)
plt.title("Simulated Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Question 9: Calculate Z-scores and visualize
# standardized data

import numpy as np
import matplotlib.pyplot as plt
z_scores = (data - np.mean(data)) / np.std(data)
plt.hist(z_scores, bins=30)
plt.title("Histogram of Z-Scores (Standardized Data)")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.show()
```
